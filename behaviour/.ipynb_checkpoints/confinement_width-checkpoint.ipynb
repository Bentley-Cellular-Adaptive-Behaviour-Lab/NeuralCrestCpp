```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
confinement_data = pd.read_csv('/Users/feldnerd/Desktop/camp-output-confexp-6Oct/simulation_summary_output_0_.csv')
```

```python
confinement_data.head()
```

```python
line = pd.pivot_table(confinement_data, 
                      values='middle_passers', 
                      index='CEwidth',
                      aggfunc = 'mean')
```

```python
line
```

```python
stdline = pd.pivot_table(confinement_data, 
                      values='middle_passers', 
                      index='CEwidth',
                      aggfunc = 'std')
```

```python
stdline
```

```python
stdline['Low'] = line['middle_passers'] - stdline['middle_passers']
stdline['High'] = line['middle_passers'] + stdline['middle_passers']
```

```python
stdline
```

```python
plt.rcParams.update({'font.size': 50})
fig, (ax) = plt.subplots(1,1, figsize=(20,10))

ax.plot(line, lw = 6, color='k')
ax.set_ylabel('Mean number of cells \nmigrating to completion', )
ax.set_xlabel('Confinement zone width (arbitrary units)', )
```

```python
fig.savefig('confinement_experiment_result.png',
            bbox_inches='tight',
            pad_inches=1)
```
