# Diversity figures

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import pandas as pd
from matplotlib import rcParams
rcParams['axes.titlepad'] = 20 
```

```python

interaction_threshold_data = np.array([[0.92,0.87], [0.10,0.74]])

cmap=matplotlib.cm.RdPu

plt.imshow(interaction_threshold_data, 
	cmap=cmap,
	origin = 'lower',
    extent = [1,2,1,2]
    )

cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 20
cbar.set_label('Mean migrating cells', rotation=270, fontsize = 12)

plt.xticks([1,2])
plt.yticks([1,2])

plt.xlabel('Leader interactionThreshold', fontsize = 12)
plt.ylabel('Follower interactionThreshold', fontsize = 12)
plt.title('interactionThreshold', fontsize = 12)


for y in range(interaction_threshold_data.shape[0]):
    for x in range(interaction_threshold_data.shape[1]):
        plt.text(x/2 +1.25, y/2 +1.25, '%.4f' % round(interaction_threshold_data[y, x],2),
                 horizontalalignment='center',
                 verticalalignment='center',)



#plt.savefig('intT.png')
```

```python

coa_data = np.array([[0.77,0.24],[0.84,0.78]])

cmap=matplotlib.cm.YlGnBu

plt.imshow(coa_data, 
	cmap=cmap,
	origin = 'lower',
    extent = [0.005,0.05,0.005,0.05]
    )
cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 20
cbar.set_label('Mean migrating cells', rotation=270, fontsize = 12)

plt.xticks([0.005,0.05])
plt.yticks([0.005,0.05])

plt.xlabel('Leader co-attraction', fontsize = 12)
plt.ylabel('Follower co-attraction', fontsize = 12)
plt.title('Co-attraction', fontsize = 12)



for y in range(coa_data.shape[0]):
    for x in range(coa_data.shape[1]):
        plt.text(x/2 +1.25, y/2 +1.25, '%.4f' % round(coa_data[y, x],2),
                 horizontalalignment='center',
                 verticalalignment='center',
                 )
        
plt.savefig('coa.png')
```
